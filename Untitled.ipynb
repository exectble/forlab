```python
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
from PIL import Image
```

```python
img = Image.open('./musk.jpg')
arr = np.array(img)

Y = []
for i in range(len(arr)):
    Yi = []
    for j in range(len(arr[i])):
        Yi.append(int(arr[i][j][0]* 0.299 + arr[i][j][1]* 0.587 + arr[i][j][2]*0.114))
    Y.append(Yi)
```

```python
plt.imshow(Y, cmap='binary_r')
```

```python
y = []
for i in Y:
    for j in i:
        y.append(j)
        
dic = Counter(y)
mylist = [key for key, val in dic.items() for _ in range(val)]
plt.hist(mylist, bins=20)
plt.show()
```

```python
def get_max_key_for_first_area():
    max_key = list(dic.keys())[list(dic.values()).index(max(dic.values()))]
    min_value = dic.get(max_key)
    keys = list(dic.keys())
    for i in range(max_key-1, 0, -1):
        if min_value >= dic[i]:
            min_value = dic[i]
            max_key = i
    return max_key

def show_plane(min_key, max_key):
    for i in range(len(arr)):
        for j in range(len(arr[i])):
            yi = arr[i][j][0]* 0.299 + arr[i][j][1]* 0.587 + arr[i][j][2]*0.114
            if yi >= min_key and yi < max_key:
                arr[i][j][0] = 255
                arr[i][j][1] = 0
                arr[i][j][2] = 0
    plt.imshow(arr)
    
def get_max_key_for_second_area():
    max_key = list(dic.keys())[list(dic.values()).index(max(dic.values()))]
    max_value = dic.get(max_key)
    min_value = 0
    avg = (max_value + min_value) / 2

    min_dif = max_value
    for i in range(255, 128, -1):
        if abs(dic.get(i) - avg) < min_dif:
            min_dif = abs(dic.get(i) - avg)
    res = 0
    for i in range(255, 128, -1):
        if dic.get(i) == avg - min_dif:
            res = i
    return res

def get_labels(max_keys):
    pixels = []
    labels = []
    for k in range(0, len(max_keys)-1):
        for i in range(len(arr)):
            for j in range(len(arr[i])):
                yi = arr[i][j][0]* 0.299 + arr[i][j][1]* 0.587 + arr[i][j][2]*0.114
                if yi >= max_keys[k] and yi < max_keys[k+1]:
                    pixels.append(arr[i][j])
                    labels.append(k)
    return pixels, labels
```

```python
max_key_plane = get_max_key_for_first_area()
max_key_sky = get_max_key_for_second_area()
max_key_smoke = max(list(dic.keys()))
pixels, labels = get_labels([0, max_key_plane, max_key_sky, max_key_smoke])
```

```python
np.savetxt('data.csv', [p for p in zip(pixels, labels)], delimiter=',', fmt='%s')
```
